```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy import stats
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
X = []
Y = []

def onclick(event):
    print('button=%d, x=%d, y=%d, xdata=%f, ydata=%f' %
          (event.button, event.x, event.y, event.xdata, event.ydata))
    X.append(event.xdata)
    Y.append(event.ydata)
    plt.plot(event.xdata, event.ydata, 'o',markersize=10)
    fig.canvas.draw()

cid = fig.canvas.mpl_connect('button_press_event', onclick)
plt.show()
```

```python
h = 1

V = srange(0,10,h/10)

F = []

for v in V: 
    # get intersection of X and ball of radius h surrounding v
    rel_X = []
    rel_Y = []
    for k in range(len(X)):
        if abs(X[k] - v) <= h:
            rel_X.append(X[k])
            rel_Y.append(Y[k])
    # calculate by r-value
    if len(rel_X) != 0:
        new_f_val = stats.pearsonr(rel_X, rel_Y)[0]
        # divide by number of points 
        #F.append(new_f_val / len(rel_X))
        F.append(new_f_val)
    else:
        F.append(0)
```

```python
fig = plt.figure()
plt.scatter(V, F)
plt.show()
```
